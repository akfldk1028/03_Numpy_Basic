## Memory Optimization

copy and view of ndarrays


copy는 a를 수정하면 b를 수정안해도 되지만 view는 a를 수정하면 b도 수정됨 그러나 메모리 최적화할때 좋다

```python
import numpy as np
a = np.arange(5)
b = a.view()
b[0] = 100
print(a)
print(b)
```

```python
import numpy as np
a = np.arange(5)
b = a[0:3]
b[...] = 10
print(a)
print(b)
```

```python
a = np.arange(5)
b = a.copy()

b[0] = 100
print(a)
print(b)
```

base of ndarrays


```python
a = np.arange(5)
b = a.copy()
c = a.view()
d = a[0:3]
print(b.base is a)
print(c.base is a)
print(d.base is a)
# copy 인지 view 인지 확인하는 방법
# if answer is True, it is copy, else it is view
```
